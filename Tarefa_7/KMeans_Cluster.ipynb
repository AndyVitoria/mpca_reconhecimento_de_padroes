```python
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
```

```python
dataset = load_iris()
X = dataset.data[:, -2:]
features_name = dataset.feature_names[-2:]
y = dataset.target
labels = dataset.target_names
```

```python
X, y, labels, features_name
```

```python
import matplotlib.pyplot as plt

for i in range(len(set(y))):
    plt.plot(X[y==i, 0], X[y==i, 1], ".", label=labels[i])
plt.xlabel(features_name[0])
plt.ylabel(features_name[1])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.plot(X[:, 0], X[:, 1], ".")
plt.xlabel(features_name[0])
plt.ylabel(features_name[1])
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

km_model = KMeans(n_clusters=3)
km_model.fit(X)
print(km_model.cluster_centers_)
```

```python
plt.scatter(X[:,0], X[:,1], marker=".")
plt.scatter(km_model.cluster_centers_[:,0], km_model.cluster_centers_[:,1], 
            marker="x", color="r", s=100, linewidths=3)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show()
```

```python
true_label_color = ['y', 'k', 'c']
for i in range(len(set(y))):
    plt.plot(X[y==i, 0], X[y==i, 1], "s", label=labels[i], color=true_label_color[i])


ypred = km_model.predict(X)

pred_label_color = ['y', 'k', 'c']
for i in range(len(set(ypred))):
    plt.plot(X[ypred==i, 0], X[ypred==i, 1], ".", label=labels[i], color=pred_label_color[i])

plt.scatter(
    km_model.cluster_centers_[:,0],
    km_model.cluster_centers_[:,1],
    marker="x",
    color="r",
    s=100,
    linewidths=3
)


plt.xlabel(features_name[0])
plt.ylabel(features_name[1])
plt.legend()
plt.show()
```
